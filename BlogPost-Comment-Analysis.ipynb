## Data Cleaning Process

Before diving into our analysis, it is crucial to prepare the data by removing any inconsistencies and filling in missing values where appropriate. This ensures the accuracy and reliability of our analysis results.

### Removing Duplicates

Duplicates in our dataset can skew the results and provide misleading insights. By removing these duplicates, we ensure that each data entry is unique and represents an actual event or observation.

### Handling Missing Values

Missing data can introduce bias and affect the validity of our statistical tests and models. Therefore, removing or imputing missing values is essential to maintain the quality of our dataset.

Below, we visualize the impact of these cleaning steps on our dataset's size, demonstrating the amount of data cleaned and retained for analysis.


### We eliminate duplicate/empty values.

#### 'test_df' is our DataFrame before cleaning

```python
initial_count = len(test_df)
```

#### Removing duplicates

```python
test_df.drop_duplicates(inplace=True)
after_duplicates_count = len(test_df)
```

#### Removing rows with any missing values

```python
test_df.dropna(inplace=True)
final_count = len(test_df)
```

#### Visualization of data cleaning effects

```python
fig, ax = plt.subplots(figsize=(8, 4))
ax.bar(['Before Cleaning', 'After Removing Duplicates', 'After Removing Missing Values'], 
       [initial_count, after_duplicates_count, final_count], color=['blue', 'orange', 'green'])
ax.set_ylabel('Number of Entries')
ax.set_title('Impact of Data Cleaning on Dataset Size')
```

#### Showing the percentage of entries removed at each step

```python
plt.figure()
sizes = [initial_count - after_duplicates_count, after_duplicates_count - final_count, final_count]
labels = ['Duplicates Removed', 'Missing Values Removed', 'Final Entries']
colors = ['red', 'orange', 'green']
explode = (0.1, 0.1, 0)  # only "explode" the first two slices

plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=90, colors=colors, explode=explode)
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.title('Percentage of Data Reduced by Cleaning Steps')

plt.show()
```

## Predictive Modeling

```python
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LinearRegression, Ridge, Lasso
from sklearn.ensemble import RandomForestRegressor, GradientBoostingRegressor
from sklearn.svm import SVR
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.neural_network import MLPRegressor
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python

X_train = train_df.iloc[:, :-1]
y_train = train_df.iloc[:, -1]
X_test = test_df.iloc[:, :-1]
y_test = test_df.iloc[:, -1]
```

```python
models = [
    ('Linear Regression', LinearRegression()),
    ('Ridge', Ridge()),
    ('Lasso', Lasso()),
    ('Random Forest', RandomForestRegressor(n_estimators=100)),
    ('Gradient Boosting', GradientBoostingRegressor(n_estimators=100)),
    ('Support Vector Regressor', SVR()),
    ('Neural Network', MLPRegressor(hidden_layer_sizes=(50, 30), max_iter=500))
]
```

```python
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
def RMSE(y_true, y_pred):
    return np.sqrt(mean_squared_error(y_true, y_pred))
```

```python
results = []

for name, model in models:
    model.fit(X_train_scaled, y_train)
    y_pred = model.predict(X_test_scaled)
    rmse = RMSE(y_test, y_pred)
    results.append((name, rmse))

```

### Convert the results to a DataFrame

```python
results_df = pd.DataFrame(results, columns=['Model', 'RMSE'])
```

### Bar graph of the RMSE of each model

```python
plt.figure(figsize=(10, 6))
sns.barplot(x='RMSE', y='Model', data=results_df.sort_values('RMSE'))
plt.xlabel('Root Mean Squared Error')
plt.ylabel('Model')
plt.title('Model Performance')
plt.show()
```

### Scatter charts to compare predictions vs actual values

```python
for name, model in models:
    y_pred = model.predict(X_test_scaled)
    
    plt.figure(figsize=(8, 5))
    sns.scatterplot(x=y_test, y=y_pred)
    plt.xlabel('True Values')
    plt.ylabel('Predictions')
    plt.title(f'{name} Predictions vs True Values')
    plt.show()
```
